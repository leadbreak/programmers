### [N으로 표현](https://school.programmers.co.kr/learn/courses/30/lessons/42895)

```python
N = 2
number = 11

dp = [set() for _ in range(9)]  # dp[0]은 사용하지 않음
for i in range(1, 9):
    # N을 i번 연속해서 사용한 숫자 (예: 5, 55, 555, ...)
    concatenated_number = int(str(N) * i)
    dp[i].add(concatenated_number)
    
    # 이전 단계의 조합을 통해 현재 단계의 숫자를 생성
    for j in range(1, i):
        for x in dp[j]:
            for y in dp[i - j]:
                dp[i].add(x + y)
                dp[i].add(x - y)
                dp[i].add(x * y)
                if y != 0:
                    dp[i].add(x // y)
    
    print(f'{dp=}')
```

```python
def solution(N, number):
    if N == number:
        return 1
    
    dp = [set() for _ in range(8+1)]
    for i in range(1, 8+1):
        dp[i].add(int(str(N)*i))
        
        for j in range(1, i):
            for x in dp[j]:
                for y in dp[i-j]:
                    if y != 0 :
                        target = [plus, minus, multiply, divide] = x+y, x-y, x*y, x // y
                        dp[i].add(divide)
                    else :
                        target = [plus, minus, multiply] = x+y, x-y, x*y
                    
                    if number in target:
                        return i
                    
                    dp[i].add(plus)
                    dp[i].add(minus)
                    dp[i].add(multiply)
                    
        if number in dp[i]:
            return i
    
    return -1
```

### [정수 삼각형](https://school.programmers.co.kr/learn/courses/30/lessons/43105)

```python
def solution(triangle):
    N = len(triangle)

    for i in range(1, N):
        triangle[i][0] += triangle[i-1][0]
        triangle[i][i] += triangle[i-1][i-1]
        
        for j in range(1, i):
            triangle[i][j] += max(triangle[i-1][j-1], triangle[i-1][j])
    
    return max(triangle[N-1])
```

### [등굣길](https://school.programmers.co.kr/learn/courses/30/lessons/42898)

```python
def solution(m, n, puddles):
    MOD = 1_000_000_007
    # 1-based indexing을 위해 (m+1) x (n+1) 크기의 DP 테이블 생성
    dp = [[0]*(n+1) for _ in range(m+1)]
    
    dp[1][1] = 1
    puddle_set = set(map(tuple, puddles))
    
    for i in range(1, m+1):
        for j in range(1, n+1):
            if (i, j) in puddle_set:
                dp[i][j] = 0
                continue
            if i == 1 and j == 1:
                continue
            dp[i][j] = (dp[i-1][j] + dp[i][j-1]) % MOD
    
    return dp[m][n]
```

### [사칙연산](https://school.programmers.co.kr/learn/courses/30/lessons/1843)

```python
def solution(arr):
    N = len(arr)//2 # 수행해야 하는 연산의 수
    history = [arr]
    for i in range(1, N):
        cur = []
        for exp in history:
            for j in range(1, len(exp), 2):
                expression = "".join(exp[j-1:j+2])
                result = eval(expression)
                new = exp[:j-1] + [str(result)] + exp[j+2:]
                cur.append(new)
        history = cur
    return max([eval("".join(x)) for x in cur])
```

```python
def solution(arr):
    # arr에서 숫자와 연산자를 분리
    nums = [int(arr[i]) for i in range(0, len(arr), 2)]
    ops = [arr[i] for i in range(1, len(arr), 2)]
    
    n = len(nums)
    
    # dp_min[i][j], dp_max[i][j]는 숫자 i부터 j까지(포함) 사용했을 때 만들 수 있는 식의 최솟값과 최댓값
    dp_min = [[0]*n for _ in range(n)]
    dp_max = [[0]*n for _ in range(n)]
    
    # 대각선(한 개의 숫자만 사용하는 구간)은 값 그대로
    for i in range(n):
        dp_min[i][i] = nums[i]
        dp_max[i][i] = nums[i]
    
    # 구간 길이(부분 문제의 크기) 2개 이상의 숫자를 다루는 경우
    for length in range(2, n+1):  # length는 구간에 포함되는 숫자의 개수
        for i in range(n-length+1):
            j = i + length - 1
            # 초기값 설정
            min_val = float('inf')
            max_val = float('-inf')
            
            # i부터 j-1까지 연산자 위치 k를 기준으로 분할
            for k in range(i, j):
                # 왼쪽 부분(i~k), 오른쪽 부분(k+1~j)을 이미 계산한 dp값 사용
                left_min, left_max = dp_min[i][k], dp_max[i][k]
                right_min, right_max = dp_min[k+1][j], dp_max[k+1][j]
                
                op = ops[k]
                
                # 가능한 조합을 모두 고려하여 최솟값/최댓값 갱신
                if op == '+':
                    val_candidates = [
                        left_min + right_min,
                        left_min + right_max,
                        left_max + right_min,
                        left_max + right_max
                    ]
                elif op == '-':
                    val_candidates = [
                        left_min - right_min,
                        left_min - right_max,
                        left_max - right_min,
                        left_max - right_max
                    ]
                
                # 구한 후보들 중 최솟값/최댓값 갱신
                min_val = min(min_val, *val_candidates)
                max_val = max(max_val, *val_candidates)
            
            dp_min[i][j] = min_val
            dp_max[i][j] = max_val
    
    # 전체 구간(0부터 n-1까지)에 대한 최댓값을 반환
    return dp_max[0][n-1]
```
