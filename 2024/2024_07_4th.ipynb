### [멀쩡한 삼각형](https://school.programmers.co.kr/learn/courses/30/lessons/62048?language=python3)

기울기를 고려해서, 한 칸 좌->우로 이동할 때 침범하는 세로 블록들의 개수를 세는 방식으로 진행  
조금이라도 침범하면 해당 사각형을 사용할 수 없기 때문에, 올림해서 사용하는 방식으로 적용

```python
import math

def solution(w,h):
    return w*h-(math.ceil(h/w)*w)
```

본 문제는 일종의 패턴이 있음. 
대각선으로 선을 그었을 때, 가로로 W개의 블록과 세로로 H개의 블록을 지나가게 되나, 이 과정에서 중복되는 블록이 생김.
정확히 격자선과 일치하는 경우.

격자선으로 돌아올 경우, 다시 이후의 패턴을 반복함.
다시 격자점으로 돌아오는 순간은 두 숫자의 최대공약수가 됨.  
이를 반영하면 아래와 같이 작성할 수 있음

```python
import math
def solution(w,h):
    return w*h-(w+h-math.gcd(w,h))
```
