- 기존 풀었던 문제 다른 방법으로 풀어보기

```python
def solution(priorities, location):
    queue =  [(i,p) for i,p in enumerate(priorities)]
    answer = 0
    while True:
        cur = queue.pop(0)
        if any(cur[1] < q[1] for q in queue):
            queue.append(cur)
        else:
            answer += 1
            if cur[0] == location:
                return answer
```

- queue 자료 구조를 고려해, deque로 풀이

```python
from collections import deque

def solution(priorities, location):
    queue = deque([(i, p) for i, p in enumerate(priorities)])
    n = 1
    while queue:
        i, p = queue.popleft()
        if any(x[1] > p for x in queue):
            queue.append((i, p))
            continue
        elif i == location:
            return n
        n += 1
```

- dfs를 이용한 풀이

```python
def solution(numbers, target):
    count = 0
    def dfs(idx, current_sum):
        nonlocal count
        if idx == len(numbers):
            if current_sum == target:
                count += 1
                print(count)
            return
        dfs(idx + 1, current_sum + numbers[idx])
        dfs(idx + 1, current_sum - numbers[idx])
    dfs(0, 0)
    return count
```

```python
numbers, target = [1, 1, 1, 1, 1], 3
solution(numbers, target)
```
