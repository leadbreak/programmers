### [비밀 코드 해독](https://school.programmers.co.kr/learn/courses/30/lessons/388352?language=python3)

```python
from itertools import combinations
from collections import deque

def solution(n, q, ans):
    targets = list(combinations([x for x in range(1, n+1)], 5))
    targets = deque([set(x) for x in targets])
    
    q = [set(y) for y in q]
    for answer, query in zip(ans, q):
        new = deque([])    
        while targets:
            numbers = targets.popleft()
            inter = len(numbers & query)
            if inter == answer:
                new.append(numbers)
        targets = new
        if not targets:
            return 0
    return len(targets)
```

- deque를 이용해 빠르게 뺐다넣긴 해도, 그 과정 자체가 반복되므로 비효율적임
- 후보 리스트에 대해 계산하는 식으로 진행해서 수정

```python
from itertools import combinations

def solution(n, q, ans):
    cnt = 0
    q = [set(x) for x in q]
    for target in combinations([y for y in range(1, n+1)], 5):
        target = set(target)
        is_valid = 1
        for answer, query in zip(ans, q):
            if len(target.intersection(query)) != answer:
                is_valid = 0
                break
        cnt += is_valid
    return cnt
```

- ans를 정렬한 다음 수행하면 더 빠르게 후보군을 좁힐 수 있지 않을까?

### [JadenCase 문자열 만들기](https://school.programmers.co.kr/learn/courses/30/lessons/12951)

```python
def solution(s):
    s = [x.lower() for x in s]
    jaden = True
    for i in range(len(s)):
        if s[i] == ' ':
            jaden = True
            continue
        elif jaden and s[i].isalpha():
            s[i] = s[i].upper()
        jaden = False
    return "".join(s)
```
