```python
from itertools import product

def solution(n, m, x, y, r, c, k):
    distance = abs(x-r) + abs(y-c)
    if distance > k or abs(distance - k) % 2 == 1:
        return 'impossible'
    
    history = [[abs(a+1-r) + abs(b+1-c) for b in range(m)] for a in range(n)]

    for combi in product(['d', 'l', 'r', 'u'], repeat=k):
        cx, cy = x, y
        possible = True
        
        for e, move in enumerate(combi, start=1):
            if move == 'd':
                cx += 1
                if cx > n: 
                    possible = False; break
            elif move == 'l':
                cy -= 1
                if cy < 1: 
                    possible = False; break
            elif move == 'r':
                cy += 1
                if cy > m: 
                    possible = False; break
            elif move == 'u':
                cx -= 1
                if cx < 1:
                    possible = False; break
            
            if history[cx-1][cy-1] > (k - e):
                possible = False
                break
        
        if possible and (cx, cy) == (r, c):
            return "".join(combi)
            
    return 'impossible'
```
