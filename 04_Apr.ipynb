### [미로탈출명령어](https://school.programmers.co.kr/learn/courses/30/lessons/150365)

```python
from itertools import product

def solution(n, m, x, y, r, c, k):
    distance = abs(x-r) + abs(y-c)
    if distance > k or abs(distance - k) % 2 == 1:
        return 'impossible'
    
    history = [[abs(a-r+1)+abs(b-c+1) for b in range(m)] for a in range(n)]

    for combi in product(['d', 'l', 'r', 'u'], repeat=k):
        cx, cy = x, y
        if combi == ('d', 'l', 'l', 'r', 'l'):
            print(combi)
            for e, move in enumerate(combi, start=1):
                print(move)
                if move == 'd':
                    cx += 1
                    if cx >= m:
                        print('scout 01', (cx, cy))
                        break
                elif move == 'u':
                    cx -= 1
                    if cx <= 0:
                        print('scout 02')
                        break
                elif move == 'l':
                    cy -= 1
                    if cy <= 0:
                        print('scout 03')
                        break
                elif move == 'r':
                    cy += 1
                    if cy >= n:
                        print('scout 04')
                        break
                print(move, (cx, cy))
                if k < (e+history[cx-1][cy-1]):
                    print('scout 05')
                    break

            print(combi, (cx,cy))
            if (cx,cy) == (r, c):
                return combi
```

```python
solution(2, 2, 1, 1, 2, 2, 2)
```
