```python
from itertools import product

def solution(n, m, x, y, r, c, k):
    distance = abs(x-r) + abs(y-c)
    if distance > k or abs(distance - k) % 2 == 1:
        return 'impossible'
    
    history = [[abs(a+1-r) + abs(b+1-c) for b in range(m)] for a in range(n)]

    for combi in product(['d', 'l', 'r', 'u'], repeat=k):
        cx, cy = x, y
        possible = True
        
        for e, move in enumerate(combi, start=1):
            if move == 'd':
                cx += 1
                if cx > n: 
                    possible = False; break
            elif move == 'l':
                cy -= 1
                if cy < 1: 
                    possible = False; break
            elif move == 'r':
                cy += 1
                if cy > m: 
                    possible = False; break
            elif move == 'u':
                cx -= 1
                if cx < 1:
                    possible = False; break
            
            if history[cx-1][cy-1] > (k - e):
                possible = False
                break
        
        if possible and (cx, cy) == (r, c):
            return "".join(combi)
            
    return 'impossible'
```

- greedy를 적용해서 모든 조합을 무식하게 탐색하지 않고 효율적으로 진행

```python
def solution(n, m, x, y, r, c, k):
    dist = abs(x - r) + abs(y - c)
    if dist > k or (k - dist) % 2 == 1:
        return "impossible"

    moves = [('d', 1, 0), ('l', 0, -1), ('r', 0, 1), ('u', -1, 0)]
    
    path = []
    curr_x, curr_y = x, y
    
    for i in range(k):
        for move, dx, dy in moves:
            nx, ny = curr_x + dx, curr_y + dy

            if 1 <= nx <= n and 1 <= ny <= m:
                remain_dist = abs(nx - r) + abs(ny - c)
                
                if remain_dist <= (k - (i + 1)):
                    path.append(move)
                    curr_x, curr_y = nx, ny
                    break 
                    
    return "".join(path)
```
