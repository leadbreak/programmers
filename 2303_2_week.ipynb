```python
def solution(input_string):
    # shortcut 1 : 한 글자인 경우
    if len(input_string)==1:
        return 'N'
    
    unique_alphas = sorted(set(input_string))
    
    # shortcut 2 : 전체 알파벳의 개수와 문자열의 길이가 같은 경우
    if len(unique_alphas) == len(input_string):
        return 'N'
    elif len(unique_alphas) == 1:
        return 'N'
    
    answer = ''
    values = {key: 0 for key in unique_alphas}
    
    input_list = list(input_string)
    before = input_list.pop()
    values[before] += 1
    for _ in range(len(input_string)-1):
        char = input_list.pop()
            
        if (before != char) and (values[char] > 0):
            answer += char
        values[char] += 1
        before = char
            
    answer = ''.join(sorted(set(answer)))
    
    return answer
```

```python
solution('aaa')
```

```python
def solution(input_string):
    # 알파벳별로 나타난 횟수와 그 알파벳이 나타나는 구분된 부분의 수를 추적하는 딕셔너리
    alpha_counts = {char: [0, 0] for char in set(input_string)}  # [등장 횟수, 구분된 부분의 수]

    prev_char = ''
    for char in input_string:
        if char != prev_char:  # 새로운 부분이 시작됨
            alpha_counts[char][1] += 1  # 구분된 부분의 수 증가
        alpha_counts[char][0] += 1  # 등장 횟수 증가
        prev_char = char

    # 외톨이 알파벳 찾기
    loners = [char for char, counts in alpha_counts.items() if counts[0] > 1 and counts[1] > 1]

    # 외톨이 알파벳이 없으면 'N' 반환
    if not loners:
        return 'N'
    else:
        # 외톨이 알파벳이 있다면 알파벳 순으로 정렬하여 반환
        return ''.join(sorted(loners))
```

```python
solution('aabcacb')
```
